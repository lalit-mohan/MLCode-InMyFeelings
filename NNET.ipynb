```python
y = y.transpose()
for column in y.columns:
    y[column].fillna('(-1,-1)', inplace=True)
print('Number of NaN values left = ', y.isnull().sum().sum())
y = y.transpose()
print('shape of X = ', X.shape, 'shape of Y = ', y.shape)
maxlen = y.shape[1]
X = pd.DataFrame(X.iloc[:,:maxlen])
y.drop('Unnamed: 0', axis=1, inplace=True)
X.drop('Unnamed: 0', axis=1, inplace=True)
print('New shape of X = ', X.shape, 'New shape of y = ', y.shape)
```

```python
y.drop('6515', axis=1, inplace=True)
```

```python
y.head()
```

```python
y.columns = y.columns.astype(int)
```

```python
for column in y.columns:
    y[column] = y[column].map(lambda x: x.lstrip('(').rstrip(')'))
    y[column] = y[column].str.replace(' ','')
```

```python
df2 = pd.DataFrame(columns=range(0, 6515))
x_series = pd.Series()
```

```python
df2.replace('-1', value=np.nan, inplace = True)
df2 = df2.astype('float64')
df2.fillna(method='ffill',axis = 1, inplace = True)
df2.fillna(method='bfill',axis = 1, inplace = True)
df2.isnull().sum().sum()
```

```python
df1.replace('-1', value=np.nan, inplace = True)
df1 = df1.astype('float64')
df1.fillna(method='ffill',axis = 1, inplace = True)
df1.fillna(method='bfill',axis = 1, inplace = True)
df1.isnull().sum().sum()
```

**Scaling/Normalization**

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
X = pd.DataFrame(scaler.fit_transform(np.array(X)))

meanX, maxX, minX = np.mean(np.array(df1)), np.max(np.array(df1)), np.min(np.array(df1))
df1 = df1.apply(lambda x: (x - np.mean(x)) / (1.5*(np.max(x) - np.min(x))))

meanY, maxY, minY = np.mean(np.array(df2)), np.max(np.array(df2)), np.min(np.array(df2))
df2 = df2.apply(lambda x: (x - np.mean(x)) / (1.5*(np.max(x) - np.min(x))))
Xcord = np.array(df1)
Ycord = np.array(df2)
```

## Generating images

```python
def generate_image(frame):
    frameCopy = np.copy(frame)
    frameWidth = frame.shape[1]
    frameHeight = frame.shape[0]
    for j in range(30):
        frame = cv2.imread('bg.jpg')
        points=[]
        for i in range(nPoints):
            cv2.circle(frame, (int(predX[j][i]), int(predY[j][i])), 5, (255, 255, 255), thickness=-1, lineType=cv2.FILLED)
            points.append((int(predX[j][i]), int(predY[j][i])))

        # Draw Skeleton
        for pair in POSE_PAIRS2:
            partA = pair[0]
            partB = pair[1]

            if points[partA] and points[partB]:
                cv2.line(frame, points[partA], points[partB], (random.choice(range(200,255)), random.choice(range(155,200)), 255), 3)
        cv2.imwrite('img-'+str(j)+'.jpg', frame)
    return None
generate_image(frame)
```
